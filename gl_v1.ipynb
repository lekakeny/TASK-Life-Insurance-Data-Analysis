```python
# Import necessary libraries

import pandas as pd
```

```python
df = pd.read_excel ('GroupLife/GL2011_2012.xlsx', engine='openpyxl')
df.head()
```

```python
# drop Null values
df1 = df.dropna(thresh = 2)
# drop empty columns
df1.dropna(axis = "columns", thresh = 10, inplace= True)
# Rename all the columns
df1.set_axis(['Name_of_Scheme', 'Policy_Number', 
              'Membership_Number', 'Rider', 'Gender', 'Policy_Type', 
              'Status', 'Level_of_Life_Cover', 'Annual_Office_Premium', 
              'Rider_Annual_Premium', 'Date_of_Birth', 'Issue_Date', 'Effective_Date', 
              'Renewal_Date', 'Exit_Date', 'Cause_of_Exit'], axis=1, inplace= True)
# Convert Date Columns into dates and treat the rest of the values as null
# df1[['Date of Birth'], ['Issue Date'], ['Effective Date'], ['Renewal Date'], ['Exit Date']] = df1[['Date of Birth', 'Issue Date', 'Effective Date', 
#      'Renewal Date', 'Exit Date']].apply(pd.to_datetime, format='%y%m%d %H:%M:%S, errors = 'coerce')
df1.iloc[:, 10:15]= df1.iloc[:, 10:15].apply(pd.to_datetime,format='%y%m%d %H:%M:%S', errors = 'coerce')                                                                                                    
                                                                                                     
                                                                                                    
                                                                                                         
```
